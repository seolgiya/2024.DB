# **경험한 서비스 (학교 서비스)**

**타겟 기능**: 지도, 수업

'지도'

**1) 시나리오 작성**

삽입: 새로운 학생 정보와 교수 정보 DB 저장

검색: 학생이 어떤 교수의 지도 or 교수가 어떤 학생을 지도하는 지 검색

삭제: 학생이나 교수가 학교에 나갈 경우 삭제



**2)SQL 쿼리:**






```python
#Install & load sqlite3
#!pip install sqlite3
import sqlite3
```

```python
# Connecting to sqlite
# connect object
conn = sqlite3.connect('SCHOOL') #DB이름: SCHOOL
```

```python
cursor_obj = conn.cursor()
```

DB 테이블 생성

**GUIDE(지도) table**

지도.png

```python
cursor_obj.execute("DROP TABLE IF EXISTS GUIDE")
```

```python
# Creating table
table = """
CREATE TABLE IF NOT EXISTS GUIDE (
    student_name VARCHAR(20),
    student_id INTEGER,
    professor_id INTEGER,
    professor_name VARCHAR(20),
    guide_id INTEGER PRIMARY KEY NOT NULL
);
"""
#SQL문을 table에 넣어둠
cursor_obj.execute(table) #위에서 정의한 sql문을 던져둠

print("Table is Ready")
```

--테이블에 데이터 **삽입**

지도.png


```python
cursor_obj.execute('''insert into GUIDE values ('KimYounghee', 202202, 8, 'JeongDawon', 1)''')
cursor_obj.execute('''insert into GUIDE values ('KimCheolsoo', 202308, 3, 'LeeJoonseok', 2)''')
cursor_obj.execute('''insert into GUIDE values ('HanWoori', 202311, 3, 'LeeJoonseok', 3)''')
cursor_obj.execute('''insert into GUIDE values ('LeeGildong', 202401, 10, 'ParkHyunjun', 4)''')
cursor_obj.execute('''insert into GUIDE values ('KimMiso', 202404, 2, 'ChoiSujin', 5)''')
```

--**검색**

전부 검색

```python
statement = '''SELECT * FROM GUIDE'''
cursor_obj.execute(statement)

print("All the data")
output_all = cursor_obj.fetchall()
for row_all in output_all:
    print(row_all)
```

학생 이름, 학번, 교수 이름만 검색

```python
statement = ''' SELECT student_name, student_id, professor_name FROM GUIDE'''
cursor_obj.execute(statement)

print("All the data")
output_all = cursor_obj.fetchall()
for row_all in output_all:
    print(row_all)
```

학번이 202401인 학생의 지도교수 이름을 검색

```python
statement = ''' SELECT professor_name FROM GUIDE WHERE student_id = 202401'''
cursor_obj.execute(statement)

print("All the data")
output_all = cursor_obj.fetchall()
for row_all in output_all:
    print(row_all)
```

--**삭제**

KimMiso 학생이 자퇴를 해 정보를 삭제 해야 하는 SQL 쿼리

```python
query_drop_table = "DELETE FROM GUIDE WHERE student_name = 'KimMiso';"
cursor_obj.execute(query_drop_table)
```

```python
statement = '''SELECT * FROM GUIDE'''
cursor_obj.execute(statement)

print("All the data")
output_all = cursor_obj.fetchall()
for row_all in output_all:
    print(row_all)
```

COURSE TABLE 생성

수업 .png






```python
cursor_obj.execute("DROP TABLE IF EXISTS COURSE")
```

```python
# Creating table
table = """
CREATE TABLE IF NOT EXISTS COURSE (
    section INTEGER,
    course_code TEXT,
    professor_id INTEGER,
    course_name VARCHAR(40),
    course_id INTEGER PRIMARY KEY NOT NULL
);
"""
#SQL문을 table에 넣어둠
cursor_obj.execute(table) #위에서 정의한 sql문을 던져둠

print("Table is Ready")
```

--테이블에 데이터 **삽입**

수업 .png

```python
cursor_obj.execute('''insert into COURSE values (1, 'hbla2017', 8, 'The_World_of_Cats', 1)''')
cursor_obj.execute('''insert into COURSE values (2, 'hbla2017', 3, 'The_World_of_Cats', 2)''')
cursor_obj.execute('''insert into COURSE values (1, 'johq1433', 2, 'Thought_and_Expression', 3)''')
cursor_obj.execute('''insert into COURSE values (1, 'hboj2001', 8, 'Creative_Minds', 4)''')
cursor_obj.execute('''insert into COURSE values (1, 'otjy1788', 10, 'Speak_with_Impact', 5)''')
```

-- **검색**

전부 검색

```python
statement = ''' SELECT * FROM COURSE'''
cursor_obj.execute(statement)
print("All the data")
output_all = cursor_obj.fetchall()
for row_all in output_all:
    print(row_all)
```

professor_id가 8인 강의 검색

```python
statement = ''' SELECT * FROM COURSE WHERE professor_id = 8'''
cursor_obj.execute(statement)
print("All the data")
output_all = cursor_obj.fetchall()
for row_all in output_all:
    print(row_all)
```

course_name이 같은 강의 검색

```python
statement = ''' SELECT * FROM COURSE WHERE course_name = 'The_World_of_Cats' '''
cursor_obj.execute(statement)
print("All the data")
output_all = cursor_obj.fetchall()
for row_all in output_all:
    print(row_all)
```

--**삭제**

Creative_Minds 강의가 폐강 되어 삭제 하는 쿼리
